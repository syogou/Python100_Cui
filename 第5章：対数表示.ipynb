```python
import math
f = lambda x: x**2 + 2 * x + 1
g = lambda x: 2**x
h = lambda x: math.log(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
domain = np.arange(1, 10, 0.1)
X = [x for x in domain]
Y1 = [f(x) for x in X]
Y2 = [g(x) for x in X]
Y3 = [h(x) for x in X]

plt.figure(figsize=(5,4))
plt.title("various functions")
plt.plot(X, Y1, label="y = f(x)")
plt.plot(X, Y2, label="y = g(x)")
plt.plot(X, Y3, label="y = h(x)" )

plt.xticks([]) #xの値を隠すために使う、初めて聞きました。

plt.grid()
plt.legend()
plt.ylabel("y")
plt.yscale("log")
plt.show()
```
